```python
#Import relevant packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Import Toy Dataset (Wine Classification Problem)
from sklearn.datasets import load_wine

wine = load_wine(as_frame = True) #as_frame returns a DataFrame object
wine.data.head()
```

```python
print(wine.DESCR)
```

```python
wine.data.info()
```

```python
wine.data.describe()
```

```python
X = wine.data

y = wine.target
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.8, random_state = 4)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

LR = LogisticRegression(max_iter=5000)
LR_model = cross_val_score(LR, X_train, y_train, cv= 5, scoring = 'accuracy')
print('LogisticRegression mean score:', LR_model.mean())

KNN = KNeighborsClassifier()
KNN_model = cross_val_score(KNN, X_train, y_train, cv= 5, scoring = 'accuracy')
print('KNeighbors mean score:', KNN_model.mean())
```

```python
KNN_list = []
for x in range(1,26):
    KNN = KNeighborsClassifier(n_neighbors = x)
    KNN_list.append(cross_val_score(KNN, X_train, y_train, cv = 8, scoring = 'accuracy').mean())
print(KNN_list)
```

```python
plt.plot( range(1,26), KNN_list)
```

```python
from sklearn import metrics
LR_model = LR.fit(X_train, y_train)
y_pred = LR_model.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```
